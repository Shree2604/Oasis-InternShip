"üöÄ Let's classify the Iris dataset using K-means clustering! We'll uncover the perfect 'k' value with the magic of the Elbow Technique. üåºüîçüìä"

```python
# Define a range of k values
k_values = range(1, 10)

# Initialize an empty list to store the sum of squared errors
sum_of_squared_errors = []

# Loop through each k value
for k in k_values:
    # Create a KMeans clustering model with 'k' clusters
    kmeans_model = KMeans(n_clusters=k)

    # Fit the model to the data using only 'petal_length' and 'petal_width' columns
    kmeans_model.fit(df[['petal_length', 'petal_width']])

    # Append the inertia (sum of squared errors) to the list
    sum_of_squared_errors.append(kmeans_model.inertia_)
```

```python
sum_of_squared_errors
```

```python
plt.xlabel('Number of Clusters (k)')
plt.ylabel("Sum of Squared Errors")
plt.plot(k_values, sum_of_squared_errors)
plt.show()
```

"üåü Let's use K-Means clustering algorithm to our dataset! üìäü§ñ‚ú®"

```python
km = KMeans(n_clusters=3,random_state=0,)
y_predicted = km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head(150)
```





Accuracy measure (üëç) (üëé) (‚úÖ) (‚ùå)

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df.species, df.cluster)
cm
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Assuming you already have the 'true_labels' and 'predicted_labels' defined.
true_labels = df.species
predicted_labels = df.cluster
class_labels = ['Setosa', 'Versicolor', 'Virginica']

# Create confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Create a figure with a grid
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)

# Show color bar
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel("Counts", rotation=-90, va="bottom")

# Set title and labels
ax.set_title('Confusion Matrix', fontsize=16)
ax.set_xlabel('Predicted label', fontsize=14)
ax.set_ylabel('True label', fontsize=14)

# Set axis ticks and labels
tick_marks = np.arange(len(class_labels))
ax.set_xticks(tick_marks)
ax.set_yticks(tick_marks)
ax.set_xticklabels(class_labels, rotation=45, ha="right", fontsize=12)
ax.set_yticklabels(class_labels, fontsize=12)

# Fill matrix with values and style text
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        text_color = "white" if cm[i, j] > cm.max() / 2 else "black"
        ax.text(j, i, str(cm[i, j]), ha='center', va='center', color=text_color, fontsize=14)

# Ensure the plot layout is tight
plt.tight_layout()

# Show the plot
plt.show()
```
